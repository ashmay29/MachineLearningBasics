# Task1:Build a linear classifier on Dataset1 using SVC.

Importing Datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.svm as svm
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.datasets import load_iris, make_classification,fetch_openml
from sklearn.exceptions import ConvergenceWarning
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
```

Making a Synthetic Dataset

```python
X,y = make_classification(n_samples=5000,n_features=10,n_classes=3,n_clusters_per_class=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15)
```

Building the SVM Model

```python
lsvc = LinearSVC(verbose=0)
print(lsvc)
LinearSVC(C=1.0, class_weight=None, dual=True, fit_intercept=True, intercept_scaling=1, loss='squared_hinge',max_iter=1000,multi_class='ovr',penalty='l2', random_state=None, tol=0.0001,verbose=0)
```

Fitting the model on the training data

```python
lsvc.fit(X_train,y_train)
score = lsvc.score(X_train,y_train)
print("Score: ",score)
```

Calculation the corss validation score

```python
cv_scores = cross_val_score(lsvc,X_train,y_train,cv=10)
print("CV Average score: %.2f" % cv_scores.mean())
```

```python
y_pred = lsvc.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
cr = classification_report(y_test,y_pred)
print(cr)
```

# Task2: Build a classifier on Dataset1 using Linear,Polynomial and RBF kernel and show the decision boundary using matplotlib.

```python
from __future__ import division, print_function
```

```python
iris = load_iris()
X = iris.data[:,:2]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
def evaluate_on_test_data(model=True):
  predictions = model.predict(X_test)
  correct_classification = 0
  for i in range(len(y_test)):
    if predictions[i]==y_test[i]:
      correct_classification += 1
  accuracy = 100*correct_classification/len(y_test)
  return accuracy
```

```python
kernels = ('linear','poly','rbf')
accuracies = []
for index,kernel in enumerate(kernels):
  model = svm.SVC(kernel=kernel)
  model.fit(X_train,y_train)
  acc = evaluate_on_test_data(model)
  accuracies.append(acc)
  print("{} % accuracy obtained with kernel = {}".format(acc, kernel))
```

```python
svc = svm.SVC(kernel='linear').fit(X_train,y_train)
rbf_svc = svm.SVC(kernel='rbf',gamma=0.7).fit(X_train,y_train)
poly_svc = svm.SVC(kernel='poly',degree=3).fit(X_train,y_train)
```

```python
h=.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

#define title for the plots
titles=['SVC with linear kernel',
        'sVC with rbf kernel',
        'SVC with polynomial kernel' ]

for i, clf in enumerate((svc,rbf_svc,poly_svc)):
  plt.figure(i)


  Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
  Z = Z.reshape(xx.shape)

  plt.contourf(xx, yy, Z, cmap=plt.cm.get_cmap('Wistia',10), alpha=0.75)
  plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.ocean)
  plt.xlabel("Sepal_Length")
  plt.ylabel("Sepal_width")
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())
  plt.xticks(())
  plt.yticks(())
  plt.title(titles[i])
plt.show()
```
