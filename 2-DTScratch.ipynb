```python
import numpy as np
def totalEntropy(df):
    target = df.keys()[-1]
    values = df[target].unique()
    entropy = 0
    for i in values:
        P = df[target].value_counts()[i] / len(df[target])
        entropy += -P * np.log2(P)
    return entropy
```

```python
target_entropy = totalEntropy(df)
print(f"The total Entropy of the target variable is:{target_entropy}")
```

```python
def attribute_entropy(df, attr):
    target = df.keys()[-1]
    target_classifiers = df[target].unique()
    attribute_classifiers = df[attr].unique()
    entropy2 = 0
    for i in attribute_classifiers:
        entropy = 0
        for j in target_classifiers:
            numerator = len(df[attr][df[attr] == i][df[target] == j])
            denominator = len(df[attr][df[attr] == i])
            P = numerator / denominator
            if P != 0:
                entropy += -P * np.log2(P)
        P2 = denominator / len(df)
        entropy2 += -P2 * entropy
    return abs(entropy2)
```

```python
def information_gain(df, attr):
    return totalEntropy(df) - attribute_entropy(df, attr)
```

```python
info_gain = information_gain(df,attr= "Temperature")
print(f"The information gain of the attribute is:{info_gain}")
```

```python
def InfoGain(df):
    IG = []
    for key in df.keys()[:-1]:
        IG.append(totalEntropy(df)-attribute_entropy(df,key))
    print(IG)
    return df.keys()[:-1][np.argmax(IG)]
```

```python
def subTable(df, attr, value):
  return df[df[attr] == value].reset_index(drop = True)
```

```python
def DecisionTree(df, tree=None):
  target = df.keys()[-1]
  SuitableAttr = InfoGain(df)
  SAvals = np.unique(df[SuitableAttr])
  if tree is None:
    tree = {}
    tree[SuitableAttr] = {}
  for value in SAvals:
    req = subTable(df, SuitableAttr , value)
    TValue,counts = np.unique(req[target],return_counts=True)
    if len(counts)==1:
      tree[SuitableAttr][value] = TValue[0]
    else:
      tree[SuitableAttr][value] = DecisionTree(req)
  return tree
```
